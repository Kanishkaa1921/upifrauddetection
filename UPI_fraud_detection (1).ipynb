```python
x = dataset[numeric_cols].values
y = dataset['isFraud'].values
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size = 0.15, random_state = 0)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

Logistic Regression (LR)

```python
from sklearn.linear_model import LogisticRegression
LR_model= LogisticRegression(random_state = 0)
```

```python
from sklearn.impute import SimpleImputer
# Check for NaN values in x_train
if np.isnan(x_train).any():
    print("NaN values found in x_train. Imputing with mean.")
    imputer = SimpleImputer(strategy='mean')
    x_train = imputer.fit_transform(x_train)

# Check for NaN values in x_test
if np.isnan(x_test).any():
    print("NaN values found in x_test. Imputing with mean.")
    imputer = SimpleImputer(strategy='mean')
    x_test = imputer.fit_transform(x_test)

#Check and impute NaN values in y_train
if np.isnan(y_train).any():
    print("NaN values found in y_train. Imputing with most frequent value.")
    imputer_y = SimpleImputer(strategy='most_frequent')
    y_train = imputer_y.fit_transform(y_train.reshape(-1, 1)).flatten()

# Check and impute NaN values in y_test
if np.isnan(y_test).any():
    print("NaN values found in y_test. Imputing with most frequent value.")
    imputer_y = SimpleImputer(strategy='most_frequent')
    y_test = imputer_y.fit_transform(y_test.reshape(-1, 1)).flatten()
```

```python
LR_model.fit(x_train, y_train) # Fit the model
y_pred = LR_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc_lr = accuracy_score(y_test,y_pred)
print(acc_lr)
```

Accuracy comparision of all models

```python
scores = [acc_lr * 100,
          acc_knn *100,
          acc_svm *100,
          acc_dt *100,
          acc_rf *100,
          acc_cnn *100]
```

```python
names = ["Logistic Regression",
         "K- Nearest neighbors",
         "Support vector Machine",
         "Decision Tree",
         "Random Forest",
         "Convolutional Neural Network"]
```

```python
df= pd.DataFrame()
df['Algorithm Nmae']= names
df['Accuracy Score(%)'] =scores
df = df.sort_values(by = 'Accuracy Score(%)',ascending = False)
df
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  counted = (series['Accuracy Score(%)']
                .value_counts()
              .reset_index(name='counts')
              .rename({'index': 'Accuracy Score(%)'}, axis=1)
              .sort_values('Accuracy Score(%)', ascending=True))
  xs = counted['Accuracy Score(%)']
  ys = counted['counts']
  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Accuracy Score(%)', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Algorithm Nmae')):
  _plot_series(series, series_name, i)
  fig.legend(title='Algorithm Nmae', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Accuracy Score(%)')
_ = plt.ylabel('count()')
```
